```python
#Soal 1
class QueueList:
    def __init__(self, kapasitas=4):
        self.queue = []
        self.waiting_list = []
        self.kapasitas = kapasitas

    def enqueue(self, tiket_id):
        if len(self.queue) >= self.kapasitas:
            print(f"Maaf, antrian penuh. {tiket_id} harus menunggu.")
            self.waiting_list.append(tiket_id)
            return

        if '7' in tiket_id:
            print(f"Pelanggan VIP {tiket_id} mendapatkan prioritas kursi!")
            self.queue.insert(0, tiket_id)
        else:
            self.queue.append(tiket_id)

    def dequeue(self):
        if not self.is_empty():
            keluar = self.queue.pop(0)
            print(f"Pelanggan {keluar} telah dilayani dan keluar dari antrian.")

            # Jika ada pelanggan di antrian tunggu, masukkan ke antrian utama
            if self.waiting_list:
                masuk = self.waiting_list.pop(0)
                print(f"Pelanggan {masuk} dari antrian tunggu masuk ke antrian utama.")
                self.queue.append(masuk)

            return keluar
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueList(kapasitas=4)

queue.enqueue("DA48")
queue.enqueue("SA49")
queue.enqueue("BA46")
queue.enqueue("AS47")  # VIP, masuk ke depan
queue.enqueue("FI45")
queue.enqueue("MU44")  # Masuk antrian tunggu

print("\nQueue setelah enqueue:", queue.queue)
print("Waiting list:", queue.waiting_list)

print("\nDequeued element:", queue.dequeue())
print("Queue setelah dequeue:", queue.queue)
print("Waiting list setelah dequeue:", queue.waiting_list)
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Soal 2
class FotokopiQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Pesanan {order_id} dengan {num_pages} halaman langsung diproses (Fast Track).")
            return

        self.queue.append((order_id, num_pages))

    def dequeue(self):
        if not self.is_empty():
            order_id, num_pages = self.queue.pop(0)

            # Jika halaman lebih dari 100, proses dalam dua tahap
            if num_pages >= 100:
                proses_halaman = 50
                sisa_halaman = num_pages - proses_halaman
                print(f"Pesanan {order_id} diproses tahap pertama ({proses_halaman} halaman). Sisa {sisa_halaman} halaman akan mengantri kembali.")
                self.enqueue(order_id, sisa_halaman)  # Masukkan kembali ke antrian paling belakang
            else:
                print(f"Pesanan {order_id} dengan {num_pages} halaman telah selesai diproses.")
            return order_id
        return None

    def front(self):
        """Menampilkan pesanan pertama dalam antrian."""
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        """Menampilkan pesanan terakhir dalam antrian."""
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        """Mengecek apakah antrian kosong."""
        return len(self.queue) == 0

    def size(self):
        """Menghitung jumlah pesanan dalam antrian."""
        return len(self.queue)

# Contoh penggunaan
queue = FotokopiQueue()
queue.enqueue("MUH48", 120)  # Diproses dalam dua tahap
queue.enqueue("MUH49", 8)    # Fast Track
queue.enqueue("MUH50", 30)
queue.enqueue("MUH51", 150)  # Diproses dalam dua tahap

print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Soal 3
class QueueList:
    def __init__(self, kapasitas=4):
        self.queue = []
        self.waiting_list = []  # Antrian tunggu jika parkiran penuh
        self.kapasitas = kapasitas

    def enqueue(self, vehicle_number, jenis_kendaraan):
        if len(self.queue) >= self.kapasitas:
            print(f"Maaf, parkiran penuh. {vehicle_number} harus menunggu.")
            self.waiting_list.append((vehicle_number, jenis_kendaraan))
            return

        if 'kendaraan_listrik' in jenis_kendaraan:
            print(f"Kendaraan listrik {vehicle_number} mendapatkan prioritas parkir!")
            self.queue.insert(0, (vehicle_number, jenis_kendaraan))
        else:
            self.queue.append((vehicle_number, jenis_kendaraan))

    def dequeue(self):
        if not self.is_empty():
            keluar = self.queue.pop(0)
            print(f"Kendaraan {keluar[0]} keluar dari parkiran.")

            # Jika ada kendaraan yang menunggu, masukkan ke parkiran
            if self.waiting_list:
                masuk = self.waiting_list.pop(0)
                print(f"Kendaraan {masuk[0]} dari antrian tunggu masuk ke parkiran.")
                self.queue.append(masuk)

            return keluar
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueList(kapasitas=4)

queue.enqueue("MA48", "kendaraan_listrik")  # Harus prioritas
queue.enqueue("MA49", "kendaraan_bensin")
queue.enqueue("MA50", "kendaraan_bensin")
queue.enqueue("MA51", "kendaraan_bensin")  # Parkiran penuh, masuk waiting list
queue.enqueue("MA52", "kendaraan_listrik")  # Masuk waiting list

print("Queue setelah enqueue:", queue.queue)
print("Waiting list:", queue.waiting_list)

print("\nDequeued element:", queue.dequeue())  # Hapus 1 kendaraan, 1 dari waiting list masuk
print("Queue setelah dequeue:", queue.queue)
print("Waiting list setelah dequeue:", queue.waiting_list)
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
class QueueList:
    def __init__(self):
        self.regular_queue = []
        self.prestasi_queue = []
        self.mixed_list = []

    def enqueue(self, student_id, category):
        if category == 'lomba_nasional':
            print(f"Mahasiswa {student_id} langsung mendapatkan sertifikat tanpa antrian.")
            return

        self.mixed_list.append((student_id, category))

    def process_queues(self):
        if len(self.mixed_list) > 10:
            while self.mixed_list:
                student_id, category = self.mixed_list.pop(0)
                if category == 'prestasi':
                    self.prestasi_queue.append(student_id)
                elif category == 'reguler':
                    self.regular_queue.append(student_id)

    def dequeue_prestasi(self):
        return self.prestasi_queue.pop(0) if self.prestasi_queue else None

    def dequeue_reguler(self):
        return self.regular_queue.pop(0) if self.regular_queue else None

    def dequeue_mixed(self):
        return self.mixed_list.pop(0) if self.mixed_list else None

    def front_prestasi(self):
        return self.prestasi_queue[0] if self.prestasi_queue else None

    def front_reguler(self):
        return self.regular_queue[0] if self.regular_queue else None

    def front_mixed(self):
        return self.mixed_list[0] if self.mixed_list else None

    def rear_prestasi(self):
        return self.prestasi_queue[-1] if self.prestasi_queue else None

    def rear_reguler(self):
        return self.regular_queue[-1] if self.regular_queue else None

    def rear_mixed(self):
        return self.mixed_list[-1] if self.mixed_list else None

    def size_prestasi(self):
        return len(self.prestasi_queue)

    def size_reguler(self):
        return len(self.regular_queue)

    def size_mixed(self):
        return len(self.mixed_list)

# Contoh penggunaan
queue = QueueList()
queue.enqueue("T48", 'lomba_nasional')  # Mahasiswa menang lomba nasional
queue.enqueue("P49", 'prestasi')  # Mahasiswa prestasi
queue.enqueue("R46", 'reguler')  # Mahasiswa reguler
queue.enqueue("P47", 'prestasi')  # Mahasiswa prestasi
queue.enqueue("A45", 'reguler')  # Mahasiswa reguler
queue.enqueue("B50", 'reguler')
queue.enqueue("C51", 'prestasi')
queue.enqueue("D52", 'reguler')
queue.enqueue("E53", 'prestasi')
queue.enqueue("F54", 'reguler')
queue.enqueue("G55", 'prestasi')
queue.enqueue("H56", 'reguler')
queue.enqueue("I57", 'prestasi')
queue.enqueue("J58", 'reguler')
queue.enqueue("K59", 'prestasi')

# Setelah semua mahasiswa masuk, proses pemisahan antrian
queue.process_queues()

print("Queue reguler:", queue.regular_queue)
print("Queue prestasi:", queue.prestasi_queue)
print("Queue campuran:", queue.mixed_list)  # Harus kosong

# Menampilkan elemen di depan dan belakang
print("Front reguler:", queue.front_reguler())
print("Front prestasi:", queue.front_prestasi())
print("Front campuran:", queue.front_mixed())
print("Rear reguler:", queue.rear_reguler())
print("Rear prestasi:", queue.rear_prestasi())
print("Rear campuran:", queue.rear_mixed())

# Menampilkan ukuran antrian
print("Queue size reguler:", queue.size_reguler())
print("Queue size prestasi:", queue.size_prestasi())
print("Queue size campuran:", queue.size_mixed())

# Menghapus elemen dari antrian dan menampilkannya
print("Dequeued reguler:", queue.dequeue_reguler())
print("Dequeued prestasi:", queue.dequeue_prestasi())
print("Dequeued campuran:", queue.dequeue_mixed())

print("Queue reguler setelah dequeue:", queue.regular_queue)
print("Queue prestasi setelah dequeue:", queue.prestasi_queue)
print("Queue campuran setelah dequeue:", queue.mixed_list)
```
