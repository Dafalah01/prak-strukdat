```python
#Soal 1
class QueueList:
    def __init__(self, kapasitas=4):
        self.queue = []
        self.waiting_list = []
        self.kapasitas = kapasitas

    def enqueue(self, tiket_id):
        if len(self.queue) >= self.kapasitas:
            print(f"Maaf, antrian penuh. {tiket_id} harus menunggu.")
            self.waiting_list.append(tiket_id)
            return

        if '7' in tiket_id:
            print(f"Pelanggan VIP {tiket_id} mendapatkan prioritas kursi!")
            self.queue.insert(0, tiket_id)
        else:
            self.queue.append(tiket_id)

    def dequeue(self):
        if not self.is_empty():
            keluar = self.queue.pop(0)
            print(f"Pelanggan {keluar} telah dilayani dan keluar dari antrian.")

            # Jika ada pelanggan di antrian tunggu, masukkan ke antrian utama
            if self.waiting_list:
                masuk = self.waiting_list.pop(0)
                print(f"Pelanggan {masuk} dari antrian tunggu masuk ke antrian utama.")
                self.queue.append(masuk)

            return keluar
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueList(kapasitas=4)

queue.enqueue("DA48")
queue.enqueue("SA49")
queue.enqueue("BA46")
queue.enqueue("AS47")  # VIP, masuk ke depan
queue.enqueue("FI45")
queue.enqueue("MU44")  # Masuk antrian tunggu

print("\nQueue setelah enqueue:", queue.queue)
print("Waiting list:", queue.waiting_list)

print("\nDequeued element:", queue.dequeue())
print("Queue setelah dequeue:", queue.queue)
print("Waiting list setelah dequeue:", queue.waiting_list)
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Soal 2
class FotokopiQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Pesanan {order_id} dengan {num_pages} halaman langsung diproses (Fast Track).")
            return

        self.queue.append((order_id, num_pages))

    def dequeue(self):
        if not self.is_empty():
            order_id, num_pages = self.queue.pop(0)

            # Jika halaman lebih dari 100, proses dalam dua tahap
            if num_pages >= 100:
                proses_halaman = 50
                sisa_halaman = num_pages - proses_halaman
                print(f"Pesanan {order_id} diproses tahap pertama ({proses_halaman} halaman). Sisa {sisa_halaman} halaman akan mengantri kembali.")
                self.enqueue(order_id, sisa_halaman)  # Masukkan kembali ke antrian paling belakang
            else:
                print(f"Pesanan {order_id} dengan {num_pages} halaman telah selesai diproses.")
            return order_id
        return None

    def front(self):
        """Menampilkan pesanan pertama dalam antrian."""
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        """Menampilkan pesanan terakhir dalam antrian."""
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        """Mengecek apakah antrian kosong."""
        return len(self.queue) == 0

    def size(self):
        """Menghitung jumlah pesanan dalam antrian."""
        return len(self.queue)

# Contoh penggunaan
queue = FotokopiQueue()
queue.enqueue("MUH48", 120)  # Diproses dalam dua tahap
queue.enqueue("MUH49", 8)    # Fast Track
queue.enqueue("MUH50", 30)
queue.enqueue("MUH51", 150)  # Diproses dalam dua tahap

print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Soal 3
class QueueList:
    def __init__(self, kapasitas=4):
        self.queue = []
        self.waiting_list = []  # Antrian tunggu jika parkiran penuh
        self.kapasitas = kapasitas

    def enqueue(self, vehicle_number, jenis_kendaraan):
        if len(self.queue) >= self.kapasitas:
            print(f"Maaf, parkiran penuh. {vehicle_number} harus menunggu.")
            self.waiting_list.append((vehicle_number, jenis_kendaraan))
            return

        if 'kendaraan_listrik' in jenis_kendaraan:
            print(f"Kendaraan listrik {vehicle_number} mendapatkan prioritas parkir!")
            self.queue.insert(0, (vehicle_number, jenis_kendaraan))
        else:
            self.queue.append((vehicle_number, jenis_kendaraan))

    def dequeue(self):
        if not self.is_empty():
            keluar = self.queue.pop(0)
            print(f"Kendaraan {keluar[0]} keluar dari parkiran.")

            # Jika ada kendaraan yang menunggu, masukkan ke parkiran
            if self.waiting_list:
                masuk = self.waiting_list.pop(0)
                print(f"Kendaraan {masuk[0]} dari antrian tunggu masuk ke parkiran.")
                self.queue.append(masuk)

            return keluar
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueList(kapasitas=4)

queue.enqueue("MA48", "kendaraan_listrik")  # Harus prioritas
queue.enqueue("MA49", "kendaraan_bensin")
queue.enqueue("MA50", "kendaraan_bensin")
queue.enqueue("MA51", "kendaraan_bensin")  # Parkiran penuh, masuk waiting list
queue.enqueue("MA52", "kendaraan_listrik")  # Masuk waiting list

print("Queue setelah enqueue:", queue.queue)
print("Waiting list:", queue.waiting_list)

print("\nDequeued element:", queue.dequeue())  # Hapus 1 kendaraan, 1 dari waiting list masuk
print("Queue setelah dequeue:", queue.queue)
print("Waiting list setelah dequeue:", queue.waiting_list)
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Soal 4
class SertifikatQueue:
    def __init__(self):
        self.reguler_queue = []  # Antrian mahasiswa reguler
        self.prestasi_queue = []  # Antrian mahasiswa berprestasi

    def enqueue(self, student_id, is_nasional):
        """Menambahkan mahasiswa ke dalam antrian berdasarkan ketentuan."""
        if is_nasional:
            print(f"{student_id} langsung mendapatkan sertifikat (Kompetisi Nasional).")
            return

        if self.size() >= 20:
            self.prestasi_queue.append(student_id)
            print(f"{student_id} masuk ke jalur mahasiswa berprestasi.")
        else:
            self.reguler_queue.append(student_id)
            print(f"{student_id} masuk ke jalur mahasiswa reguler.")

    def dequeue(self):
        """Melayani mahasiswa berdasarkan prioritas antrean."""
        if self.prestasi_queue:
            return self.prestasi_queue.pop(0)  # Prioritaskan mahasiswa berprestasi
        if self.reguler_queue:
            return self.reguler_queue.pop(0)
        return None

    def front(self):
        """Menampilkan mahasiswa pertama dalam antrian."""
        if self.prestasi_queue:
            return self.prestasi_queue[0]
        if self.reguler_queue:
            return self.reguler_queue[0]
        return None

    def rear(self):
        """Menampilkan mahasiswa terakhir dalam antrian."""
        if self.reguler_queue:
            return self.reguler_queue[-1]
        if self.prestasi_queue:
            return self.prestasi_queue[-1]
        return None

    def is_empty(self):
        """Mengecek apakah antrian kosong."""
        return len(self.reguler_queue) + len(self.prestasi_queue) == 0

    def size(self):
        """Menghitung jumlah mahasiswa dalam antrian."""
        return len(self.reguler_queue) + len(self.prestasi_queue)

# Contoh penggunaan
queue = SertifikatQueue()
queue.enqueue("T48", False)  # Mahasiswa reguler
queue.enqueue("T47", True)   # Mahasiswa langsung dapat sertifikat
queue.enqueue("T46", False)
queue.enqueue("T45", False)

print("Queue reguler:", queue.reguler_queue)
print("Queue prestasi:", queue.prestasi_queue)
print("Dequeued:", queue.dequeue())
print("Front:", queue.front())
print("Rear:", queue.rear())
print("Size:", queue.size())
```
